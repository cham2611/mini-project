```python
import pandas as pd
data=pd.read_csv('/content/cars.csv')
```

```python
data_ex2=data.copy()
```

```python
data_ex2.info()
```

```python
result=['옵션_선루프', '옵션_파노라마선루프', '옵션_열선앞', '옵션_열선뒤', '옵션_전방센서', '옵션_후방센서',
 '옵션_전방캠', '옵션_후방캠', '옵션_어라운드뷰', '옵션_네비순정', '소유자변경횟수', '사고상세_전손',
 '사고상세_침수전손', '사고상세_침수분손', '사고상세_도난', '보험_내차피해(횟수)', '보험_내차피해(가격)',
 '사고상세_타차가해(횟수)', '보험_타차피해(가격)']

data_preprocessing = data_ex2.dropna(subset=result)
```

```python
data_preprocessing.info()
```

```python
option_name=['옵션_선루프','옵션_파노라마선루프','옵션_열선앞','옵션_열선뒤','옵션_전방센서','옵션_후방센서','옵션_전방캠','옵션_후방캠','옵션_어라운드뷰','옵션_네비순정']

for i in option_name:
    data_preprocessing[i] = data_preprocessing[i].str.replace('유', '1').str.replace('무', '0')

    # 문자열을 정수로 변환
    data_preprocessing[i] = data_preprocessing[i].astype(int)

name=['소유자변경횟수','사고상세_전손','사고상세_침수전손','사고상세_침수분손','사고상세_도난','보험_내차피해(횟수)','사고상세_타차가해(횟수)']
for i in name:
    data_preprocessing[i] = data_preprocessing[i].astype(int)

name_ex1=['보험_타차피해(가격)','보험_내차피해(가격)']

for i in name_ex1:
    data_preprocessing[i] = data_preprocessing[i].str.replace(',','')

    # 문자열을 정수로 변환
    data_preprocessing[i] = data_preprocessing[i].astype(int)

data_preprocessing['주행거리']=data_preprocessing['주행거리'].str.replace(',','').str.replace('km','').str.replace('mi','')
data_preprocessing['주행거리']=data_preprocessing['주행거리'].astype(int)

data_preprocessing['가격']=data_preprocessing['가격'].str.replace(',','').str.replace('만원','').str.replace('보류','0').str.replace('계약','0').str.replace('[0]','0').str.replace('원','')
data_preprocessing['가격']=data_preprocessing['가격'].astype(int)

data_preprocessing['배기량'] = data_preprocessing['배기량'].str.replace(r',', '', regex=True).str.replace(r'\s*cc\s*\(\d+마력\)', '', regex=True).str.replace('cc ()','').str.replace('cc (-마력)','')
data_preprocessing['배기량']=data_preprocessing['배기량'].astype(int)

data_preprocessing['연식']=data_preprocessing['연식'].str.extract(r'(\d{4})\.(\d{2})')[0] + data_preprocessing['연식'].str.extract(r'(\d{4})\.(\d{2})')[1]
data_preprocessing['연식'] = data_preprocessing['연식'].apply(
    lambda x: (2022 - int(x[:4])) * 12 + int(x[5:7])
)

data_preprocessing['보험이력등록']= data_preprocessing['보험이력등록'].str.replace('등록','1').str.replace('미등록','0').str.replace('미1','0')
data_preprocessing['보험이력등록']=data_preprocessing['보험이력등록'].astype(int)

data_preprocessing['신차대비가격']= data_preprocessing['신차대비가격'].str.replace('준비중','0').str.replace('소유','0').str.replace('운용','0').str.replace('렌트승계','0').str.replace('반납','0')
data_preprocessing['신차대비가격']=data_preprocessing['신차대비가격'].astype(int)
```

```python
data_preprocessing['연료'] = data_preprocessing['연료'].replace({
    '가솔린 하이브리드': '기타',
    '가솔린/LPG겸용': '기타',
    '전기': '기타',
    '하이브리드(가솔린/전기)': '기타'
})
data_preprocessing['연료'].value_counts()
```

```python
data_preprocessing['보증정보'] = data_preprocessing['보증정보'].replace({
    '만료': '0',
    '불가':'0',
    '정보없음': '0',
    '60개월 / 100,000km': '1',
    '60개월 / 120,000km': '1',
    '가능': '1'
})
data_preprocessing['보증정보'].value_counts()
```

```python
def update_name(name):
    name = name.strip()  # 공백 제거
    if "현대" in name:
        return "현대"
    elif "기아" in name:
        return "기아"
    elif "제네시스" in name:
        return "제네시스"
    return name

data_preprocessing['이름'] = data_preprocessing['이름'].apply(update_name)
```

```python
data_preprocessing['이름'].value_counts()
```

```python
data_preprocessing['색상'] = data_preprocessing['색상'].replace({
    '검정색': '검정색',
    '검정투톤': '검정색',
    '검정색 (오로라 블랙)': '검정색',
    '검정색 (오로라 블랙 펄)': '검정색',
    '흰색': '흰색',
    '흰색투톤': '흰색',
    '흰색 (스노우 화이트 펄)': '흰색',
    '흰색 (랩핑)': '흰색',
    '흰색 (스노우 화이트)': '흰색',
    '흰색 (스노우 화이트펄)': '흰색',
    '흰색 (블랙 무광랩핑)': '흰색',
    '기타색상': '기타',
    '진회색': '기타',
    '청색': '기타',
    '은색': '기타',
    '회색': '기타',
    '노란색': '기타',
    '갈색': '기타',
    '빨간색': '기타',
    '진주색': '기타',
    '진청색': '기타',
    '남색': '기타',
    '파란색': '기타',
    '자주색': '기타',
    '녹색': '기타',
    '연두색': '기타',
    '베이지색': '기타',
    '보라색': '기타',
    '은색 (블랙)': '기타',
    '진주투톤': '기타',
    '주황색': '기타',
    '흰색 (핑크)': '흰색',
    '흰색 (레드)': '흰색',
    '청옥색': '기타',
    '은하늘색': '기타',
    '회색 (코스모 그레이)': '기타',
    '은색투톤': '기타'
})

data_preprocessing['색상'].value_counts()
```

```python
data_preprocessing.drop(['링크','차량번호','보험이력등록'],axis=1, inplace=True)
```

```python
name1=['사고상세_전손','사고상세_침수전손','사고상세_침수분손','사고상세_도난']
name2=['옵션_선루프','옵션_파노라마선루프','옵션_열선앞','옵션_열선뒤','옵션_전방센서','옵션_후방센서','옵션_전방캠','옵션_후방캠','옵션_어라운드뷰','옵션_네비순정']
data_preprocessing['사고 합']= data_preprocessing[name1].sum(axis=1)
data_preprocessing['옵션 합']= data_preprocessing[name2].sum(axis=1)
```

```python
data_preprocessing.drop(['Unnamed: 0.1', 'Unnamed: 0'], axis=1, inplace=True)
data_preprocessing.drop(name1, axis=1, inplace=True)
data_preprocessing.drop(name2, axis=1, inplace=True)
```

```python
data_oh = data_preprocessing.copy()
data_oh
```

```python
data_oh = pd.get_dummies(data_oh, columns = ['연료','색상','이름','보증정보'], dtype=int)
data_oh
```
