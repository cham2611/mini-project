```python
import yfinance as yf  # 야후 파이낸스(Yahoo Finance)의 차트 주가 정보를 가져온다.
import pandas as pd
import datetime
import matplotlib.pyplot as plt

pd.set_option('display.precision', 4)  # 소수점 아래 4자리까지만 출력할 수 있도록 pandas가 설정


start = datetime.datetime(2022, 6, 3)
end = datetime.datetime(2023, 6, 3)

dfn = yf.download("^KS11", start=start, end=end)  # ^KS11에 해당하는 코스피지수 데이터를 start와 end 기간 내에서 가져옴
df = dfn.dropna()  # 결측값 제거

df['Price'] = 0   # 새로운 열 price 추가
for i in range(0,(len(df)-1)):  # df의 각 행에 접근
    if df['Close'][i] < df['Close'][i+1]:   # Close값이 다음 행의 Close값보다 작으면 1 아니면 0
        df['Price'][i] = 1                  # 즉, 주가 상승 = 1, 하락 = 0
    else:
        df['Price'][i] = 0

df.to_csv('kospi주가데이터.csv')
df
```

```python
import pandas as pd
from urllib import parse
import requests
from bs4 import BeautifulSoup

# 팍스넷 뉴스 크롤링 함수
def paxnet_news_title(dates):
    base_url = 'http://www.paxnet.co.kr/news/much?newsSetId=4667&currentPageNo={}&genDate={}&objId=N4667'
    headers = { 'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/84.0.4147.89 Safari/537.36' }

    for date in dates:
        for page in range(1, 3): # 1~2 페이지까지 반복
            url = base_url.format(page, date)   # 페이지와 날짜를 포맷팅해서 요청할 URL을 생성
            res = requests.get(url, headers=headers)   # get요청
            if res.status_code == 200:  # 응답 상태 코드가 200이면
                soup = BeautifulSoup(res.text)    # BeautifulSoup을 사용하여 HTML 페이지를 파싱
                t_list = soup.select('ul.thumb-list li')  # 뉴스 제목이 포함된 요소 선택
                for title in t_list:  # title_list에 대해 반복
                    try:
                        news_title = title.select_one('dt').text.strip()   # 각 요소에서 뉴스 제목 추출
                        result_list.append([news_title])    # result_list에 추가
                    except:
                        error_cnt += 1   # 뉴스 제목이 존재하지 않을 경우

result_list = []
error_cnt = 0

paxnet_news_title(price_0)
t_df_0 = pd.DataFrame(result_list, columns=['뉴스제목'])
t_df_0['주가변동'] = 0

result_list = []
error_cnt = 0

paxnet_news_title(price_1)
t_df_1 = pd.DataFrame(result_list, columns=['뉴스제목'])
t_df_1['주가변동'] = 1


```

```python
# 뉴스 제목과 주가변동 값을 담는 데이터 프레임 생성
t_df = pd.concat([t_df_0, t_df_1])
t_df.to_csv('팍스넷_뉴스타이틀.csv', index=False, encoding='utf-8')
t_df
```
